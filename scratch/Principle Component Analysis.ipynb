```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121, aspect="equal")
ax2 = fig.add_subplot(122, aspect="equal")
```

```python
data = np.random.multivariate_normal([1, 1.2], [[0.4, 1], [0.025, 0.2]], size=10000)
ax1.clear()
ax1.scatter(data[:, 0], data[:, 1], s=0.5, lw='0', color="red", alpha="0.4")
ax1.set_xlim(-6, 6)
ax1.set_ylim(-6, 6)

# now we can try to find the principle components

centered = data - data.mean(axis=0)
mat_cov = np.dot(centered.T, centered)

U, s, V = np.linalg.svd(mat_cov)
```

```python
ax2.clear()
ax2.scatter(data.dot(U[:, 0]), data.dot(U[:, 1]), lw='0', color="#23aaff", alpha="0.3")
```

```python
fig = plt.figure(figsize=(12, 5))
ax3 = fig.add_subplot(121, projection="3d", aspect="equal")
ax4 = fig.add_subplot(122, projection="3d", aspect="equal")
```

```python
data = np.random.multivariate_normal([1, 1.2, 0.4], [[0.4, 1, 0], [0.025, 0.2, 0], [0.2, 0.4, 0.15]], size=10000)
ax3.clear()
ax3.scatter3D(data[:, 0], data[:, 1], data[:, 2], s=0.5, color="#23aaff")
ax3.set_xlim(-6, 6)
ax3.set_ylim(-6, 6)
ax3.set_zlim(-6, 6)

# now we can try to find the principle components

centered = data - data.mean(axis=0)
mat_cov = np.dot(centered.T, centered)

U, s, V = np.linalg.svd(mat_cov)

ax4.clear()
ax4.scatter(data.dot(U[:, 0]), data.dot(U[:, 1]), 0, s=0.5, lw='0', color="#ff0000")
ax4.set_xlim(-6, 6)
ax4.set_ylim(-6, 6)
ax4.set_zlim(-6, 6);
```
